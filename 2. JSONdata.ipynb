# JSON Data
JavaScript Object Notation plays a crucial role in web development, particularly in the context of data exchange between web applications and servers.
The simplicity, readability, and versatility of JSON have made it the standard choice for data interchange in modern web development.

JSON data is typically received as a string, especially when retrieving data from external sources like web APIs. To work with this data in JavaScript, you need to convert it from a string into a JavaScript object. **`JSON.parse()`** performs this conversion

```javascript
const getData = (callback) => {
    const request = new XMLHttpRequest();

    request.addEventListener('readystatechange', () => {
        if (request.readyState === 4 && request.status === 200) {
            const data = JSON.parse(request.responseText);   // converting JSON to JS object
            callback(undefined, data);
        }
        else if (request.readyState === 4) {

            callback('could not fetch the data.', undefined);
        }

    })
    request.open('GET', 'https://jsonplaceholder.typicode.com/todos/')
    request.send(); }

    getData((err, data) => {
        if (err) {
            console.log(err);
        }
        else {
            console.log('Response Data:\n',data);
        }

    })
```
