# HTTP Requests

- Make HTTP requests to get data from another server.
- Let's make these requests to API endpoints.
  Using `{JSON}-Placeholder` a Free fake API for testing

```javascript
const request = new XMLHttpRequest();
request.open('GET','https://jsonplaceholder.typicode.com/todos/')
request.send();
```

out: 

#### Track the progress and access the data responded with.
This can be achieved by using a specific event listener named `"readystatechange"`



```javascript
const request = new XMLHttpRequest();

request.addEventListener('readystatechange', () => {
    if (request.readyState === 4 ) {    //checking if state4, i.e operation is completed
        console.log('Below is the response Data');
        console.log(request.responseText);
    }

})

request.open('GET','https://jsonplaceholder.typicode.com/todos/')
request.send();
```

out: <br>


#### Now even the operation is successful, there is a possibility that the request may be fail for any reason.
To check this too we can use `.status`.


```javascript
const request = new XMLHttpRequest();

request.addEventListener('readystatechange', () => {
    if (request.readyState === 4 && request.status === 200) { //check if request HTTP code is 200(OK)
        console.log('Below is the response Data');
        console.log(request.responseText); // contains response Data
    }
    else if (request.readyState === 4){
        console.log('Request Failed!');    // log a fail message to console
    }
})

request.open('GET','https://jsonplaceholder.typicode.com/todoss/')
request.send();
```
